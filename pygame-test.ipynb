Extremely simple run through of pygame controller (without game)

```python
import sys
import pygame
from pygame.locals import *
pygame.init()
#clock = pygame.time.Clock() # I don't know if you need this to just print controller input
```

```python
joysticks = [pygame.joystick.Joystick(i) for i in range(pygame.joystick.get_count())]
#print(joysticks)
for joystick in joysticks:
    print(joystick.get_name())
pygame.joystick.init()
```

### Print events:

```python
while True:
    for event in pygame.event.get():
        if event.type == pygame.JOYAXISMOTION:
            print(event)
        if event.type == pygame.JOYBUTTONDOWN:
            print(event)
```

### Print Button Presses:

```python
while True:
    for event in pygame.event.get():
        if event.type == pygame.JOYBUTTONDOWN:
            if event.button == 4:
                print('LB')
            if event.button == 5:
                print('RB')
            if event.button == 8:
                print('L JOY PRESS')
            if event.button == 9:
                print('R JOY PRESS')
            if event.button == 3:
                print('Y')
            if event.button == 2:
                print('X')
            if event.button == 0:
                print('A')
            if event.button == 1:
                print('B')
```

You can add JOYHATMOTION in a similar fashion if you want to use that too...

### Print Joystick Values:

```python
while True:
    for event in pygame.event.get():
        if event.type == pygame.JOYAXISMOTION:
            print(event.axis, event.value)
```

### Allow for reconnection in case controller gets disconnected:

```python
while True:
    for event in pygame.event.get():
        if event.type == pygame.JOYDEVICEADDED:
            joysticks = [pygame.joystick.Joystick(i) for i in range(pygame.joystick.get_count())]
            print('DEVICE ADDED')
        if event.type == pygame.JOYDEVICEREMOVED:
            joysticks = [pygame.joystick.Joystick(i) for i in range(pygame.joystick.get_count())]
            print('DEVICE REMOVED')
```
